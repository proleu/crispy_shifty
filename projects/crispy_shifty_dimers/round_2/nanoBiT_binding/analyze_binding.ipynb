```python
import matplotlib.pyplot as plt
import numpy as np

from scipy.optimize import curve_fit
l0 = 1.0 # nM
def binding_curve(s0, p, Kd): # units of luminescence, s0 = total constant component concentration (nM), p = conversion factor from luminescence to concentration of complex, Kd = dissociation constant (nM)
    b = s0 + l0 + Kd
    c = s0*l0
    return (b - np.sqrt(b**2 - 4*c))/(2*p)

fig, axs = plt.subplots(nrows=2, ncols=4, figsize=(16,8))

for i, (combo_name, start_conc, ax) in enumerate(zip(combo_names, start_concs, axs.flatten())):

    start_ind = i*12 + (i//8+1)*2
    binding_data = binding_df.iloc[5, start_ind:start_ind+12]
    s0 = np.array([start_conc*1000/(4**i) for i in range(12)])
    try:
        (p, Kd), pcov = curve_fit(binding_curve, s0, binding_data, bounds=(0,np.inf))
        ax.plot(s0, binding_data*p/l0, '.', label='Data')
        s0_plot = np.logspace(np.log10(s0[-1]), np.log10(s0[0]))
        ax.plot(s0_plot, binding_curve(s0_plot, p, Kd)*p/l0, '-', label='Fit', zorder=-1)
    except RuntimeError:
        print(f"{combo_name} fit failed")
    ax.set_xscale('log')
    ax.set_xlim(0.001, 100000)
    ax.set_ylim(0, 1)
    ax.set_xlabel('Total Conc (nM)')
    ax.set_ylabel('Fraction in Complex')
    ax.set_title(f'{combo_name} Kd = {Kd:.01f}nM')

fig.tight_layout()
fig.savefig("20220614_CSD_binding_fits.png", dpi=300)
```
