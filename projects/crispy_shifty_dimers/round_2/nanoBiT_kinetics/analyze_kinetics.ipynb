```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(12,6))

l0 = 0.1 # nM
def binding_curve(s0, p, Kd): # units of luminescence, s0 = total constant component concentration (nM), p = conversion factor from luminescence to concentration of complex, Kd = dissociation constant (nM)
    b = s0 + l0 + Kd
    c = s0*l0
    return (b - np.sqrt(b**2 - 4*c))/(2*p)

Kds = []

for i, (combo_data, fits, ax) in enumerate(zip(combo_datas, fits_all, axs)):

    concs = np.array([c for c in fits.keys()])
    sats = np.array([fit[0] for fit in fits.values()])

    sns.scatterplot(x=concs/1000000000, y=sats, ax=ax)

    try:
        (p, Kd), pcov = curve_fit(binding_curve, concs, sats, bounds=(0,np.inf))
        conc_plot = np.logspace(np.log10(concs[-1]), np.log10(concs[0]))
        sns.lineplot(x=conc_plot/1000000000, y=binding_curve(conc_plot, p, Kd), zorder=-1, ax=ax)#, legend=None)
        ax.text(0.05, 0.95, f'Kd = {Kd/1000000000:.00e} M\n[{combo_data[2]}] = 1e-10 M', horizontalalignment='left', verticalalignment='top', transform=ax.transAxes)
        Kds.append(Kd/1000000000)
    except RuntimeError:
        print(f"{combo_data} fit failed")

    ax.set_xscale('log')
    ax.set_xlabel(f'[{combo_data[1]}] (M)')
    ax.set_ylabel('Saturation Luminescence')

fig.tight_layout()
fig.patch.set_alpha(0)
fig.savefig("CSD20f3_kinetics_sat_conc.png", dpi=300)
```
